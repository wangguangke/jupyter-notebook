```python
# 导入需要用到的库
import pandas as pd
import numpy as np
```

```python
#创建特征列表
column_names = ['Sample code number', 'Clump Thickness', 'Unigormity og Cell Size',
                'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single EpithelitalCellSize',
                'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses','Class']
```

```python
#读取数据
data=pd.read_csv('breast-cancer-wisconsin.data', names =column_names)
data.describe()
```

```python
#对数据进行简单预处理
#将？替换为标准缺失值
data = data.replace(to_replace = '?', value = np.nan)
#去掉带有缺失值的数据
data = data.dropna(how = 'any')
#再次对数据进行描述
data.describe()
```

```python
#对数据进行分割
from sklearn.model_selection import train_test_split
#75%的数据用做训练集，25%的数据用做测试集
X_train, X_test, y_train, y_test = train_test_split(data[column_names[1:10]], 			    data[column_names[10]], test_size = 0.25, 			    random_state = 33)

#查验训练样本的数量和类别分布
y_train.value_counts()

#查验测试样本的数量和类别分布
y_test.value_counts()
```

```python
#导入需要用到的算法
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
#对数据进行标准化处理
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
#初始化logistic回归器和SGDClassifier
lr = LogisticRegression()
```

```python
#训练模型
lr.fit(X_train, y_train)
```

```python
#预测
lr_y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
#使用logistic自带的评分函数
print('Accuracy of LR Classifier:', lr.score(X_test, y_test))
```

```python
#其他指标
print(classification_report(y_test, lr_y_pred, target_names = ['Benign', 'Malignant']))
```
